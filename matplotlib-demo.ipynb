```python
data = [7150, 5420, 4130, 3190, 2940, 1137]
labels = ['SAMAUNG','HUAWEI','APPLE','XIAOMI','OPPO','OTHER']
explode = [0,0.1,0,0,0,0]
autopct = '%1.1f%%'

plt.pie(data, labels=labels, explode=explode, autopct=autopct)
```
